```python
df.head()
```

```python
df = df.dropna()
```

```python
one_hot_encoded_data = pd.get_dummies(df,drop_first=True, dtype = "int")
one_hot_encoded_data.head(5)
```

```python
one_hot_encoded_data.shape
```

```python
X1 = one_hot_encoded_data.drop("Yıl",axis=1)  #modelin tahmin edeceği değer çıkartılırr.
Y1 = one_hot_encoded_data[["Kasıtlı_Hektar", "İhmal_Hektar", "Doğal_Hektar", "Bilinmeyen_Hektar"]]
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X1_train = ss.fit_transform(X1_train)
X1_test = ss.transform(X1_test) #burayı fit etmemelisiniz. Aksi taktirde model test verilerindeki verileri de ezberleyecektir.
#Bu doğru bir yaklaşım değildir. Çünkü veri sızıntısına ( Data leak) sebebiyet verir.
```

```python
X1_train.shape
```

```python
X1_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

model1 = LinearRegression()
model1.fit(X1_train, Y1_train)

Y1_pred = model1.predict(X1_test)
mse_stage1 = mean_squared_error(Y1_test, Y1_pred)

# 2030 yılı için yangın sayısı tahmini
model1.fit(df[["Yıl"]], df[["Kasıtlı_Hektar", "İhmal_Hektar", "Doğal_Hektar", "Bilinmeyen_Hektar"]])

predicted_counts = model1.predict(pd.DataFrame({"Yıl": [2030]}))

etiketler = ["Kasıtlı", "İhmal", "Doğal", "Bilinmeyen"]
for isim, tahmin in zip(etiketler, predicted_counts[0]):
    print(f"- {isim} Yangın Tahmini: {tahmin:.0f} hektar")
```

```python
# Gerçek veriyi al
veri_df = df[["Yıl", "Kasıtlı_Hektar", "İhmal_Hektar", "Doğal_Hektar", "Bilinmeyen_Hektar"]].dropna()

# Yıl aralığı ve tahmin yapılacak gelecek yıllar
years = list(veri_df["Yıl"].values)
future_years = list(range(max(years)+1, 2031))
full_years = years + future_years

# Her kolon için model eğit ve tahmin üret
tahmin_dict = {}

for column in ["Kasıtlı_Hektar", "İhmal_Hektar", "Doğal_Hektar", "Bilinmeyen_Hektar"]:
    model = LinearRegression()
    model.fit(veri_df[["Yıl"]], veri_df[[column]])

    tahmin = model.predict(pd.DataFrame({"Yıl": full_years}))
    tahmin_dict[column] = tahmin.flatten()

# Grafik çizimi
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))

for column in tahmin_dict:
    plt.plot(full_years, tahmin_dict[column], label=column)

plt.axvline(x=max(years), color='gray', linestyle='--', label="2024 sonrası tahmin")
plt.title("Kasıtlı, İhmal, Doğal ve Bilinmeyen Yangınlara Göre Yanan Alan Tahmini (1987–2030)")
plt.xlabel("Yıl")
plt.ylabel("Yanan Alan (Hektar)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Subnational Treecover Loss**

```python

df=pd.read_csv(file2, encoding="windows-1254", sep=";")
df.head()
```

```python
yenibasliklar = {
    "ï»¿country": "Ülke",
    "subnational1": "Bölge",
    "threshold": "Eşik_Değeri",
    "area_ha": "Alan_ha",
    "extent_2000_ha": "Alan_2000_ha",
    "extent_2010_ha": "Alan_2010_ha",
    "gain_2000-2012_ha": "Kazanç_2000_2012_ha",
    # Sonraki yıllara ait kayıp kolonları için de şöyle:
    "tc_loss_ha_2001": "Orman_Kaybı_2001_ha",
    "tc_loss_ha_2002": "Orman_Kaybı_2002_ha",
    # ...
    "tc_loss_ha_2024": "Orman_Kaybı_2024_ha"
}
# tc_loss_ha_YYYY kolonlarını otomatik çevir
for year in range(2001, 2025):
    eski = f"tc_loss_ha_{year}"
    yeni = f"Orman_Kaybı_{year}_ha"
    if eski in df.columns:
        yenibasliklar[eski] = yeni


df = df.rename(columns=yenibasliklar)

df.head()
```

```python
print(df.columns.tolist())
```

```python
#Yıllara Göre Orman Kaybı Toplamı (Türkiye Genelinde)

import matplotlib.pyplot as plt

# Yıllık orman kaybı kolonlarını seç
yillar = list(range(2001, 2025))
kolonlar = [f"Orman_Kaybı_{y}_ha" for y in yillar]

# Türkiye genelindeki yıllık toplam kayıp
yillik_toplam = df[kolonlar].sum()

plt.figure(figsize=(12,6))
plt.plot(yillar, yillik_toplam, marker='o')
plt.title("Türkiye Genelinde Yıllara Göre Orman Kaybı (ha)")
plt.xlabel("Yıl")
plt.ylabel("Orman Kaybı (hektar)")
plt.grid(True)
plt.show()

```

```python
# Bölgelere Göre Toplam Orman Kaybı (2001-2024)
 # Bölge bazında toplam kayıp (tüm yıllar toplamı)
df["Toplam_Orman_Kaybı_ha"] = df[kolonlar].sum(axis=1)

# Bölgeye göre toplam kayıp
bölge_kayip = df.groupby("Bölge")["Toplam_Orman_Kaybı_ha"].sum().sort_values(ascending=False)

plt.figure(figsize=(19,9))
bölge_kayip.plot(kind='bar')
plt.title("Bölgelere Göre Toplam Orman Kaybı (2001-2024)")
plt.xlabel("Bölge")
plt.ylabel("Toplam Orman Kaybı (hektar)")
plt.xticks(rotation=90)
plt.show()
```
